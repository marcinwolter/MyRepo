# **Babylonian method of square root calculation**

```
# Babylonian method
# Importing math to use sqrt function
import math

def f(S, x):
   return x*x -S

# Re-writing f(x)=0 to x = g(x)
def g(S, x):
   return (x+S/x)/2.

# Implementing Fixed Point Iteration Method
def fixedPointIteration(S, x0, e, N):
   print('\n\n*** FIXED POINT ITERATION ***')
   step = 1
   flag = 1
   condition = True
   while condition:
      x1 = g(S, x0)
      print('Iteration-%d, x1 = %0.8f and f(x1) = %0.8f' % (step, x1, f(S, x1)))
      x0 = x1
      nn.append(step)
      result.append(x1)
      step = step + 1
      
      if step > N:
         flag=0
         break
      condition = abs(f(S, x1)) > e
   if flag==1:
       print('\nRequired root is: %0.8f' % x1)
   else:
       print('\nNot Convergent.')


# Input Section
S = input('Enter S (sqrt(S) will be found): ')
x0 = input('Enter Guess: ')
e = input('Tolerable Error: ')
N = input('Maximum Step: ')
# Converting x0 and e to float
S = float(S)
x0 = float(x0)
e = float(e)
# Converting N to integer
N = int(N)
#Note: You can combine above three section like this
# x0 = float(input('Enter Guess: '))
# e = float(input('Tolerable Error: '))
# N = int(input('Maximum Step: '))
# Starting fixed point iteration

result = []
nn = []

fixedPointIteration(S,x0,e,N)
```

Plot the resulting function

```
# Import pyplot as plt
import matplotlib.pyplot as plt
import operator

# define a figure (1 row, 1 column)
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 12))

# fill numpy array with true value of sqrt(S)
sqres = result
sqres = [math.sqrt(S)] * len(result)



ax[0].plot(nn,result,lw=2)
ax[0].plot(nn,sqres,lw=3)
ax[0].set_title("Approximation of sqrt")

#log scale
plt.yscale('linear')
ax[0].grid()

diff = list(map(operator.sub, result, sqres))
ax[1].plot(nn,diff,lw=2)
ax[1].set_title("Difference approx-true")

#log scale
plt.yscale('log')
ax[1].grid()

plt.show()
```
